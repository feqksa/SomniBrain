```python
import numpy as np
import matplotlib.pyplot as plt

# Intensity calculations
def I_fun(I0_val, r_val,):
  I_dr = []
  I_dr.append(I0_val)
  for r in range(1, len(r_val)):
    I_dr.append(I0_val / (2 * np.pi * (r_val[r] ** 2)))
  return np.array(I_dr)

# Opitcal power calculations

optical_power = 400 #mW
area = 2 * np.pi *(2.2 * 10 **(-1))**2
I0 = optical_power / area
i = 0

# Worst case scenario :
# Received power by the head
r_vals = np.linspace(1, 30, num = 300)
r_axis = np.linspace(0, 30, num = 300)

I = I_fun(I0, r_vals)

# Plot not accounting for the angle
print(I[0])
print(I[2])
plt.figure(1)
plt.plot(r_axis, I)
plt.xlabel("Distance (cm)")
plt.ylabel("Light intensity (mW/cm^2)")
```

```python
# For which distance do we have I = 50 mW/cm^2 ?

for i in range(1, len(r_vals)):
  if I[i] >= 20:
    print(r_vals[i-1])
```

```python
# Taking the angle into account

h = 20 # cm, height between monkey and optic fiber
head_d = 12 # cm, diameter of monkey's head

teta_max = np.arctan(head_d / (2 * h))
teta_max_deg = teta_max * 360 / (2 * np.pi)
print(teta_max_deg)
```

```python
# For a 16° angle, we have 15% of the initial intensity

I_min = 0.15 * I

plt.figure(1)
plt.plot(r_axis, I)
plt.xlabel("Distance (cm)")
plt.ylabel("Light intensity (mW/cm^2)")
```

```python
# For which distance do we have Imin = 50 mW/cm^2 ?

for i in range(1, len(r_vals)):
  if I_min[i] > 50:
    print(r_vals[i-1])
```
